```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
name = ["A","B","C","D","E"]
```

```python
salary = np.random.randint(10000, 70000,5)
```

```python
color1 = ["g","b","k","#c4ae70","#ae70e0"]
```

```python
plt.bar(name, salary, color = color1)
plt.xlabel("Name")
plt.ylabel("Salary")
plt.title("Name vs Salary")
plt.show()
```
